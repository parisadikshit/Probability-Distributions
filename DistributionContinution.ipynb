# Bernoulli Distribution

>  It is a discrete type of probability distribution. In Bernoulli Distribution,the random variable can take only two values : 0 and 1 i.e.(success or failure). For bernoulli distribution,applies certain conditions:
1. only 2 possible outcomes success or failure
2. P(success) = p while P(failure)=1-p
3. Generally, we take P(success)>P(failure)

```python
from scipy.stats import bernoulli
```

```python
data_bernoulli = bernoulli.rvs(size=1000,p=0.6)
```

```python
data_bernoulli
```

```python
data_bernoulli.shape
```

```python
#plotting with random variables
ax = sns.distplot(
    data_bernoulli,
    kde = True,
    hist_kws={'linewidth':20,'alpha':1},
    color = 'skyblue',
    
)
ax.set(xlabel='Bernoulli Distribution',ylabel='Frequency')
```

```python
# Defining the sucess of probability
p = 0.6
```

> Expected value E[X] is to specify the likelihood of the favoured event

> $ E[X] = \sum x*P(X=x)   $   
In Bernoulli,it becomes equal to p.

```python
# mean
mean = bernoulli.mean(p)
print("Mean:",mean)
```

```python
# median 
median = bernoulli.median(p)
print("Median:",median)
```

> $ Variance = E[X]^2 -E[X]^2 $    
Variance can be given as p(1-p)

```python
# variance
variance = bernoulli.var(p)
print("Variance:",variance)
```

$ Standard Deviation = \sqrt{Variance} $

```python
# standard deviation
std = bernoulli.std(p)
print("Standard deviation:",std)
```

```python
# probability mass function
print("PMF:",bernoulli.pmf(data_bernoulli,p))
```

```python
#plotting scatter plot of PMF
plt.scatter(data_bernoulli,bernoulli.pmf(data_bernoulli,p),label="PMF")
plt.title('Scatter Plot of Probability Mass Function for p=0.6')
plt.xlabel('Data points')
plt.ylabel('Probability')
plt.legend(loc='upper left')
```

```python
# PLotting PMF
plt.bar(data_bernoulli,bernoulli.pmf(data_bernoulli,p),width=0.1,color=['green','red'],label="PMF")
plt.title("Probability Mass Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")
```

```python
# Plotting CDF
plt.scatter(data_bernoulli,bernoulli.cdf(data_bernoulli,p),label="CDF")
plt.title("Scatter Plot of CDF at p=0.6")
plt.xlabel("Data Points")
plt.ylabel("Probability")
plt.legend(loc='upper left')
```

```python
plt.bar(data_bernoulli,bernoulli.cdf(data_bernoulli,p),width=0.1,color=['green','red'],label="CDF")
plt.title("Cummulative Distribution Function")
plt.xlabel("Data Points")
plt.ylabel("Probability")
colors = {'Success':'green', 'Failure':'red'}  
```

# numpy.random.seed()

> numpy.random.seed is for pseudo random numbers,the word pseudo random here means as appearing to be random but in fact predetermined. This is used to generate the same exact random numbers and recreate the exact same set of pseudo random numbers. 

> Running the 2 lines below give the same output no matter how many times run.For the same seed value,the same set of random numbers are created. If you provide the same seed,you'll get the same output.

```python
np.random.seed(0)
np.random.randint(low=1,high=10,size=5)
```

> On the other hand runnning only the line without the line with the seed,will give random numbers everytime it got run.
